```python
# 챗봇 응답
def request_answer_by_chatbot( text ):
    # 1. 토크나이저를 이용하여 인코딩
    # </s><usr>질문<sys>
    bos = [ tokenizer.bos_token_id ]
    sentence = tokenizer.encode( f'{tokenizer.decode(2)}{text}{tokenizer.decode(4)}' )
    sentence = bos + sentence
    # print( sentence )

    # 2. sentence를 텐서로 변경
    sent_tensor = tf.convert_to_tensor([sentence])    # 학습의 n개의 문장으로 학습했으므로 형식을 2D로 맞춘다

    # 3. 모델에 데이터 주입 -> 예측(=답변을 생성) 수행, 답변 50토큰까지 진행 => 패딩사이즈에 제한걸림
    #    텐서로 응답
    answer = model.generate(
                sent_tensor,
                do_sample=True, #샘플링 전략 사용
                max_length=50, # 최대 디코딩 길이는 50
                top_k=10, # 확률 순위가 10위 밖인 토큰은 샘플링에서 제외 (속도와 성능면에서 10으로 설정함)
                top_p=0.95, # 누적 확률이 95%인 후보집합에서만 생성
                num_return_sequences=3 #3개의 결과를 디코딩해낸다
            )
    # print( answer )
    
    # 4. 실데이터만 추출해서, 리스트로 변환 -> 디코딩 수행
    answer_list = []
    for a in answer:
        answer = tokenizer.decode( a.numpy().tolist() )
        # 5. 실제응답한 내용만 추출
        # print(answer)
        answer_list.append(answer.split(tokenizer.decode(4))[-1].split(tokenizer.decode(1))[0].strip())

    return answer_list
```

```python
# 감정분류

def predict(predict_sentence):

    text = predict_sentence
    # predict
    preds_list = text_classifier(text)[0]

    sorted_preds_list = sorted(preds_list, key=lambda x: x['score'], reverse=True)
    emotion_num = sorted_preds_list[0]['label'] # label
    # print(emotion_num)
    predicted_score = sorted_preds_list[1]['score'] # score
    # print(predicted_score)

    emotion_num = int(emotion_num.replace("LABEL_", "").strip())

    if emotion_num == 0:
        emotion_result = '불안'
    elif emotion_num == 1:
        emotion_result = '당황'
    elif emotion_num == 2:
        emotion_result = '분노'
    elif emotion_num == 3:
        emotion_result = '슬픔'
    elif emotion_num == 4:
        emotion_result = '기쁨'
    elif emotion_num == 5:
        emotion_result = '상처'

    #print('-' * 50)
    # print(f'emotion: {emotion_result} {predicted_score*100:.2f}%')

    # label, 확률값 리턴
    return emotion_result, predicted_score
```

```python
# 말투 스타일 리스트
style_list = ['초등학생', '중학생', '할아버지', '할머니']
```

```python
# 말투 변환
def generate_text(pipe, text, target_style, num_return_sequences=5, max_length=60):
  target_style_name = style_list[target_style]
  text = f"{target_style_name} 말투로 변환:{text}"
  out = pipe(text, num_return_sequences=num_return_sequences, max_length=max_length)
  return [x['generated_text'] for x in out]
```

```python
# make_cartoon
def make_cartoon(img, size=450, style="Hayao"):
    model = Transformer()
    model.load_state_dict(torch.load(os.path.join('/content/drive/MyDrive/CartoonGAN/사전학습모델/pretrained_model', style + '_net_G_float.pth')))
    model.eval()
    # print('{} Model loaded!'.format(style))
    img_size = 256
    T = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize(img_size, 2),
    transforms.ToTensor()])
 
    img_input = T(img).unsqueeze(0)
    img_input = -1 + 2 * img_input
    
    img_output = model(img_input)
 
    img_output = (img_output.squeeze().detach().numpy() + 1.) / 2.
    img_output = img_output.transpose([1, 2, 0])

    return img_output
    
    # plt_imshow([style], [img_output], figsize=(16, 10))
```

## main 함수



이미지 업로드해서 StarGAN/data/0번에 저장되게 해서 StarGAN으로 표정 변환 후 Cart toon으로 변환 시켜놓기

그 뒤에 채팅하면서 이미지도 imshow로 보여주기

=> Happy(기쁨), Sad(슬픔+상처), Angry(분노),Surprise(당황), Fear(불안) 


```python
def main():
  # 설정 
  # 이미지 설정
  print("캐릭터 이미지를 지정해주세요!")
  f = files.upload()
  print('이미지 변환 중입니다...')

  # 이미지 변환
  img = cv2.imread('/content/test.jpg', cv2.IMREAD_COLOR)

  ## 얼굴 crop -> 256X256 resize
  # grayscale로 변환
  image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
  # print(image)

  # Detect faces
  faces = face_cascade.detectMultiScale(image, 1.1, 4)

  # Draw rectangle around the faces and crop the faces
  for (x, y, w, h) in faces:
      # cv2.rectangle(image, (x, y), (x+w, y+h), (0, 0, 255), 2)
      crop_img = img[y:y + h, x:x + w]
      crop_img = cv2.resize(crop_img, dsize=(256, 256), interpolation=cv2.INTER_CUBIC)
      break

  # 생성 모델에 넣기
  image_dic = gan(crop_img)
  # print(image_dic)
  output.clear()
  plt_imshow('Jieun', img=[image_dic['기본']], figsize=(8, 5))

  # 캐릭터 선택
  character = int(input("   원하시는 캐릭터 말투를 선택해주세요! : 1.초등학생, 2.중학생, 3.할아버지, 4.할머니 *숫자만 써주세요 : "))
  print('\n')
  print("   설정이 완료되었습니다. 채팅을 시작합니다.")
  print('\n')

  end = 1

  # 감정 지수 리스트
  emotion_list = {
      "불안" : 0,
      "당황" : 0,
      "분노" : 0,
      "슬픔" : 0,
      "기쁨" : 0,
      "상처" : 0,
  }
  
  # 챗봇
  while end == 1 :
      print("   * (0. 채팅종료)")
      sentence = input("    하고싶은 말을 입력해주세요 : ")
      # 0을 누르면 나가짐
      if sentence == "0" :
          output.clear()
          emotion_result = max(emotion_list,key=emotion_list.get)
          print('*' * 50)
          print('-' * 50)
          print("당신은 오늘 " + emotion_result + " 가득하네요.")
          print(text_list[emotion_result])
          print('-' * 50)
          print('*' * 50)
          break

      user_emotion, user_p = predict(sentence)
      print(f'    label:{user_emotion}, 예측 확률:{user_p * 100:.2f}%')
      print("\n")
      answer_3 = request_answer_by_chatbot(sentence)
      
      max_p = 0
      max_s = " 가득하네요."  # 임의 답변
      max_l = user_emotion

      # 기쁨은 기쁨으로만 답변
      if user_emotion == "기쁨" :
          for a in answer_3:
              l, p = predict(a)
              # print(a)
              if l == "기쁨":
                  if p > max_p:
                      max_p = p
                      max_s = a
                      max_l = l
      else :
          for a in answer_3:
              l, p = predict(a)
              # print(a)
              if l != "기쁨":
                  if p > max_p:
                      max_p = p
                      max_s = a
                      max_l = l

      if max_s == " 가득하네요.":
          max_s = user_emotion + max_s

      max_s = generate_text(nlg_pipeline, max_s, character-1, num_return_sequences=1, max_length=50)[0]
      plt_imshow('Jieun', img=[image_dic[max_l]], figsize=(8, 5))
      print(f'    답변 : {max_s}')
      print(f'    label:{max_l}, 예측 확률:{max_p * 100:.2f}%')
      emotion_list[max_l] += 1
      print("\n")
      time.sleep(5)
      output.clear()
```

```python
main()
```
